```python
import pandas as pd
from pprint import pprint
```

```python
MAIN_FOLDER = "/home/arman/it/AI_work/analyze_data/perfume"
```

```python
data = pd.read_csv(f"{MAIN_FOLDER}/dataset/perfume.csv")
print(data.shape)
data.head()
```

```python
full_duplicates = data[data.duplicated(keep=False)]
print(f"Количество полных дубликатов: {full_duplicates.shape[0]}")
print(data.shape)
full_duplicates.head()
```

```python
data = data[data.duplicated(keep="first") == False]  # noqa: E712
print(data.shape)
data.head()
```

```python
text_cols = ["brand", "title", "type", "itemLocation", "priceWithCurrency", "availableText"]
for col in text_cols:
    data[col] = data[col].astype("string").str.strip()

if data["is_man"].dtype != "bool":
    data["is_man"] = data["is_man"].astype(str).str.strip().str.lower().map({"true": True, "false": False})

for col in ["price", "available", "sold"]:
    data[col] = pd.to_numeric(data[col], errors="coerce")

data["title"] = data["title"].str.replace(r"\s+", " ", regex=True)

data.head()
```

```python
country_map = {
    "usa": "United States",
    "u.s.a": "United States",
    "us": "United States",
    "united states": "United States",
    "hongkong": "Hong Kong",
    "hong kong": "Hong Kong",
}

country_raw = data["itemLocation"].astype("string").str.split(",").str[-1].str.strip().str.lower()

data["country"] = country_raw.replace(country_map)
data["country"] = data["country"].str.title()

print(data["country"].value_counts())
data[["itemLocation", "country"]].head()
```

```python
data.to_csv(f"{MAIN_FOLDER}/dataset/cleaned_perfume.csv", index=False)
```
