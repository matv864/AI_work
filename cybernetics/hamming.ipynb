```python
from typing import Any, Literal
```

```python
class HammingComfort:
    @classmethod
    def encode(cls, record: Any) -> list[Literal[0,1]]:
        match record:
            case int():
                return cls.encode_bits(list(map(int, format(record, "b"))))
            case str():
                return cls.encode_bits(list(map(int, record)))
            case list():
                return cls.encode_bits(record)
            case _:
                return []
    
    @staticmethod
    def encode_bits(record: list[Literal[0,1]]) -> list[Literal[0,1]]:
        mixed_record = []
        cur_control_level = 1
        cur_record_i = 0
        cur_i = 1
        while True:
            if cur_control_level == cur_i:
                mixed_record.append(0)  # control bit
                cur_control_level *= 2
            else:
                if cur_record_i >= len(record):
                    break
                mixed_record.append(record[cur_record_i])
                cur_record_i += 1
            cur_i += 1
        # print(mixed_record)
        cur_level = 1
        while cur_level < len(mixed_record):
            start = cur_level - 1
            res_for_i = 0
            for i in range(start, len(mixed_record), cur_level*2):
                res_for_i = ((sum(mixed_record[i:i+cur_level])) + res_for_i) % 2
                # print(cur_level, mixed_record[i:i+cur_level], [t+1 for t in range(i, i+cur_level)])
            mixed_record[start] = res_for_i

            cur_level *= 2

        print(mixed_record)
        return mixed_record

    @staticmethod
    def fix_bits(record: list[Literal[0,1]]):
        variant = 0

        cur_level = 1
        while cur_level < len(record):
            start = cur_level - 1
            res_for_i = 0
            for i in range(start, len(record), cur_level*2):
                res_for_i = ((sum(record[i:i+cur_level])) + res_for_i) % 2
                # print(cur_level, record[i:i+cur_level], [t+1 for t in range(i, i+cur_level)])
            if record[start] != (res_for_i - record[start] + 2) % 2:
                variant += cur_level

            cur_level *= 2

        if variant:
            variant -= 1
            record[variant] = 1 - record[variant]
        return record

        

        

```

```python
t = HammingComfort.encode(154)
# t[0] = 1 - t[0]
t[2] = 1 - t[2]
print(t)
HammingComfort.fix_bits(t)
print(t)
print("=====")
t = HammingComfort.encode("0100010000111101")
t[10] = 1 - t[10]
print(t)
HammingComfort.fix_bits(t)
print(t)
print("----")
```
