```python
import numpy as np
import random
from pprint import pprint
from typing import Literal
```

```python
BOOK_FOLDER = "/home/arman/it/AI_work/crypto/data"
book_1 = f"{BOOK_FOLDER}/1984.txt"
book_2 = f"{BOOK_FOLDER}/Animal_Farm.txt"
book_3 = f"{BOOK_FOLDER}/Keep_the_Aspidistra_Flying.txt"
```

```python
all_letters = {chr(i) for i in range(ord("a"), ord("z")+1)}
```

```python
with open(book_1) as F:
    book_1_text = F.read()

with open(book_2) as F:
    book_2_text = F.read()

with open(book_3) as F:
    book_3_text = F.read()

book_texts = [book_1_text, book_2_text, book_3_text]
```

```python
for i in range(len(book_texts)):
    book_texts[i] = book_texts[i].lower().replace("\n", " ")
```

```python
new_book_texts = []
for book_text in book_texts:
    new_book_text = []
    for i in book_text:
        if ("a" <= i <= "z") or i in " ":
            new_book_text.append(i.lower())
    new_book_texts.append("".join(new_book_text))
book_texts = new_book_texts
```

```python
book_texts[0][:20]
```

```python
books_words = []
for book_text in book_texts:
    books_words.append(book_text.split())
books_words[0][:5]
```

```python
books_digit_to_letter = []
books_letter_to_digit = []
for book_words in books_words:
    books_digit_to_letter_item = {}
    books_letter_to_digit_item = {}
    for word_i in range(len(book_words)):
        books_digit_to_letter_item[word_i] = book_words[word_i][0]
        books_letter_to_digit_item[book_words[word_i][0]] = books_letter_to_digit_item.get(book_words[word_i][0], []) + [word_i]
    books_digit_to_letter.append(books_digit_to_letter_item)
    books_letter_to_digit.append(books_letter_to_digit_item)
```

```python
for i in range(len(book_texts)):
    print(f"в книге {i} c {len(books_words[i])} словами есть {len(books_letter_to_digit[i])} букв, с которых начинается хотя бы одно слово, нет этих букв:", end=" ")
    print(all_letters - set(books_letter_to_digit[i].keys()))
```

```python
def crypt(text: str, book_num: Literal[1,2,3]=1) -> list[int]:
    used = {letter: 0 for letter in all_letters}
    words = ("".join(filter(lambda x: "a" <= x <= "z" or x == " ", text.lower().replace("\n", " ")))).split()
    digit_words = []
    for word in words:
        record = []
        for let in word:
            allow_digits = books_letter_to_digit[book_num][let]
            need_digit = allow_digits[used[let]%len(allow_digits)]
            used[let] += 1
            record.append(need_digit)
        digit_words.append(record)
    result = []
    for digit_word in digit_words:
        result += digit_word
        result.append(-1)
    return result
```

```python
def decrypt(cipher: list[int], book_num: Literal[1,2,3]) -> list[int]:
    result = ""
    for digit in cipher:
        if digit == -1:
            result += " "
            continue
        result += books_digit_to_letter[book_num][digit]
    return result
```
