```python
import math
import random
```

```python
alph = "АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ"
# alph = "абвгдежзийклмнопрстуфхцчшщъыьэюя"
# alph = "АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯабвгдежзийклмнопрстуфхцчшщъыьэюя"

len_alph = len(alph)
```

```python
class Vernam:
    def __init__(self):
        self.mod = len(alph)
        self.alph_dict = {a: (ord(a)-ord(alph[0])) for a in alph}
        self.alph_dict_rev = {v: k for k, v in self.alph_dict.items()}
 
    def generate_key(self, n: int) -> str:
        return "".join([self.alph_dict_rev[random.randint(0, self.mod-1)] for _ in range(n)])
    
    def encrypt(self, data: str, visual: bool = True) -> str:
        key = self.generate_key(len(data))
        crypted_data = ""
        for data_sym, key_sym in zip(data, key):
            crypted_data += self.alph_dict_rev[(self.alph_dict[data_sym] ^ self.alph_dict[key_sym]) % self.mod]
        if visual:
            print(f"{key} - ключ")
        return crypted_data

```

```python
data = "ЧТОНИБУДЬ"

cypher = Vernam()
print(f"{data} - исходное")
encoded_data = cypher.encrypt(data)
print(f"{encoded_data} - зашифровано")
```
