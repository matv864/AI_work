```python
from typing import Sequence, Optional, Any

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer

import pymorphy3

NLTK_DIR = "C:\\nltk_data"
# nltk.data.path.clear()
# nltk.data.path.append(NLTK_DIR)

nltk.download('all', download_dir=NLTK_DIR)

```

Написать программу на языке программирования Python, реализующую:

1) Предварительную обработку текста, включающую:
    - Перевод текста в нижний регистр
    - Удаление знаков препинания
    - Удаление стоп-слов
    - Удаление лишних символов
    - Стемминг и лемматизацию


```python
def process_text(text: str) -> Sequence[str]:
    text = text.lower()
    text = "".join(filter(
        lambda sign: sign in " " or 'а' <= sign <= 'я' or 'А' <= sign <= 'Я' or 'a' <= sign <= 'z' or 'A' <= sign <= 'Z', 
        text
    ))
    word_tokens = word_tokenize(text, language='russian')
    stop_words = set(stopwords.words('russian'))
    filtered_words = filter(lambda word: word not in stop_words, word_tokens)

    return filtered_words

```

```python
def stemm_words(words: list[str]) -> list[str]:
    stemmer = SnowballStemmer("russian")
    stemmed_words = [stemmer.stem(word) for word in words]
    return stemmed_words

```

```python
def lemmatization(words: list[str]) -> list[str]:
    result: list[str] = []
    morph = pymorphy3.MorphAnalyzer(lang='ru')
    for word in words:
        word = word.strip()
        word = morph.normal_forms(word)[0]
        result.append(word)
    
    return result
```

```python
def create_bow(text: str) -> dict[str, str]:
    stemmer = SnowballStemmer("russian")

    words = process_text(text)
    words = list(words)

    bag_of_word = dict()
    for word in words:
        stemmed_word = stemmer.stem(word)
        counter = 0
        for check_word in words:
            counter += int(stemmed_word in check_word)
        bag_of_word[stemmed_word] = bag_of_word.get(stemmed_word, 0) + 1
    return bag_of_word
```

```python
print(stemm_words(process_text("на столе стол ложка с другой ложкой")))
print(lemmatization(process_text("на столе стол ложка с другой ложкой")))
```
